```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
data_dir = 'D:/new new data'
```

```python
image_height, image_width = 48, 48
batch_size = 32
num_epochs = 10
num_classes = 7
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  
)
```

```python
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(image_height, image_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  
)
```

```python
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(image_height, image_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  
)
```

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(image_height, image_width, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(train_generator,
                    steps_per_epoch=train_generator.samples // batch_size,
                    epochs=num_epochs,
                    validation_data=validation_generator,
                    validation_steps=validation_generator.samples // batch_size)
```

```python
test_data_dir = 'D:/new new data'
```

```python
test_generator = datagen.flow_from_directory(
    test_data_dir,
    target_size=(image_height, image_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  
)
```

```python
test_loss, test_accuracy = model.evaluate(test_generator, steps=test_generator.samples // batch_size)
```

```python
from tensorflow.keras.preprocessing import image

img_path = 'D:/MMi/fear/S019-1088.jpg'  
img = image.load_img(img_path, target_size=(image_height, image_width))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.  

prediction = model.predict(img_array)

predicted_class = np.argmax(prediction)

class_names = ['Angry', 'contempt', 'disgust', 'fear', 'happy', 'sadness', 'surprise']

print(f"The predicted emotion for the image is: {class_names[predicted_class]}")
```
